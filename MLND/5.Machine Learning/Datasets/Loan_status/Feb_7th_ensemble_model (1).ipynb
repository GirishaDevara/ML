```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('loan_status.csv')
data.head()
```

```python
print(data.columns)
print(data.shape)
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data.shape
```

```python
y=data.Loan_Status
data = data.drop(axis=1,columns='Loan_Status')
print(data.shape)
print(y.shape)
```

```python
data.dtypes
```

```python
data.describe(include='all')
```

```python
data.Dependents.head(10)
```

```python
x = data.drop(axis=1,columns='Loan_ID')
x = pd.get_dummies(x) 
x.shape
```

```python
y=y.map(arg={'Y':1,'N':0})
y.head(5)
```

```python
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
print(x_train.shape,y_train.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
model1 = KNeighborsClassifier()
model1 = model1.fit(x_train,y_train)
train_pred1 = model1.predict(x_train)
test_pred1 = model1.predict(x_test)
print(accuracy_score(train_pred1,y_train))
print(accuracy_score(test_pred1,y_test))
```

```python
model1 = KNeighborsClassifier(3)
model1 = model1.fit(x_train,y_train)
train_pred1 = model1.predict(x_train)
test_pred1 = model1.predict(x_test)
print(accuracy_score(train_pred1,y_train))
print(accuracy_score(test_pred1,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
model2.fit(x_train,y_train)
train_pred2 = model2.predict(x_train)
test_pred2 = model2.predict(x_test)
print(accuracy_score(train_pred2,y_train))
print(accuracy_score(test_pred2,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier(max_depth=3)
model2.fit(x_train,y_train)
train_pred2 = model2.predict(x_train)
test_pred2 = model2.predict(x_test)
print(accuracy_score(train_pred2,y_train))
print(accuracy_score(test_pred2,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
model3 = LogisticRegression()
model3.fit(x_train,y_train)
train_pred3 = model3.predict(x_train)
test_pred3 = model3.predict(x_test)
print(accuracy_score(train_pred3,y_train))
print(accuracy_score(test_pred3,y_test))
```

```python
from sklearn.naive_bayes import BernoulliNB
model4 = BernoulliNB()
model4.fit(x_train,y_train)
train_pred4 = model4.predict(x_train)
test_pred4 = model4.predict(x_test)
print(accuracy_score(train_pred4,y_train))
print(accuracy_score(test_pred4,y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
model4 = GaussianNB()
model4.fit(x_train,y_train)
train_pred4 = model4.predict(x_train)
test_pred4 = model4.predict(x_test)
print(accuracy_score(train_pred4,y_train))
print(accuracy_score(test_pred4,y_test))
```
